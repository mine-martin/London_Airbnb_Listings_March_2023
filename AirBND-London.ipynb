```python
# import pandas library
import pandas as pd
```

```python
# read in the csv file as a pandas dataframe
airBNB = pd.read_csv("listings.csv")
```

```python
# expore the data
airBNB.info()
```

```python
airBNB.shape
```

```python
# Create a new DataFrame with only the desired columns
selected_columns = ['name', 'host_name', 'latitude', 'longitude', 'property_type', 'room_type', 'price', 'minimum_nights', 'maximum_nights', 'availability_30', 'availability_60', 'availability_90', 'availability_365', 'number_of_reviews', 'review_scores_rating']
new_airBNB = airBNB.loc[:, selected_columns]

# expore the new df data
new_airBNB.info()

new_airBNB.shape
```

```python
# check the head of the df
new_airBNB.head()
```

```python
# add new columns and data to the table
selected_column = ['name', 'host_name', 'latitude', 'longitude', 'host_since', 'host_location', 'host_acceptance_rate',  'property_type', 'room_type', 'price', 'minimum_nights', 'maximum_nights', 'availability_30', 'availability_365', 'number_of_reviews', 'review_scores_rating']
newAirBNB = airBNB.loc[:, selected_column]

# expore the new df data
newAirBNB.info()

newAirBNB.shape

newAirBNB.head()
```
